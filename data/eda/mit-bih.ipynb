N = Normal beat (normal sinus rhythm).

A = Atrial premature beat (aka PAC, premature atrial contraction).

V = Premature ventricular contraction (PVC).

Each of these annotations is tied to a sample index (time point `annotation.sample`), which is the reference location of the R-peak. These are the “ground truth” QRS locations you use for error checking (±75 ms window).

`+` = Rhythm change marker. This isn’t a beat. It’s inserted in the annotations to signal that the rhythm classification (e.g., “atrial fibrillation,” “sinus rhythm,” etc.) changes at that point.

For detection accuracy, all beat types (N, A, V, etc.) are considered true QRS complexes. Non-beat symbols (like +) are ignored in the detection task, since they don’t correspond to a QRS.

You line up your detected R-peaks against the indices of all annotated beats. Check if each detection falls within ±75 ms of an annotation. The annotation’s letter code doesn’t matter for QRS detection — just the fact that a beat was present.

Double check durations and sampling frequency

```python
# Make a dictionary where the key is (shape, fs)
from collections import defaultdict
duration_sampling_freq_counter = defaultdict(int)

data_dir = '../raw/mit-bih-arrhythmia-database-1.0.0'
for fname in os.listdir(data_dir):
    if fname.endswith('.hea'):
        record_name = fname.replace('.hea', '')
        record = wfdb.rdrecord(os.path.join(data_dir, record_name))
        signal_shape = record.p_signal.shape
        fs = record.fs
        key = signal_shape + (fs,)
        duration_sampling_freq_counter[key] += 1

print(duration_sampling_freq_counter)
```

What to do about multiple annotations for a record? annotations can have different file paths than `id`.atr

```python
annotation = wfdb.rdann('../raw/mit-bih-arrhythmia-database-1.0.0/108', 'at_')
# Access annotation data
print(annotation.sample)      # Annotation sample indices
print(annotation.symbol)      # Annotation symbols

annotation2 = wfdb.rdann('../raw/mit-bih-arrhythmia-database-1.0.0/108', 'atr')
# Access annotation data
print(annotation2.sample)      # Annotation sample indices
print(annotation2.symbol)      # Annotation symbols

def find_all_diff_indices(list1, list2):
    diff_indices = []
    min_length = min(len(list1), len(list2))
    for i in range(min_length):
        if list1[i] != list2[i]:
            diff_indices.append(i)
    # If one list is longer, the remaining elements are also considered differences
    if len(list1) > len(list2):
        for i in range(min_length, len(list1)):
            diff_indices.append(i)
    elif len(list2) > len(list1):
        for i in range(min_length, len(list2)):
            diff_indices.append(i)
    return diff_indices


assert (annotation.sample == annotation2.sample).all()
try:
    assert annotation.symbol == annotation2.symbol
except AssertionError as e:
    print(f"Assertion failed: {e}")

diff_indices = find_all_diff_indices(annotation.symbol, annotation2.symbol)
print(diff_indices)
print(f"{annotation.symbol[227:230]} vs {annotation2.symbol[227:230]}")
print(f"{annotation.symbol[1761:1764]} vs {annotation2.symbol[1761:1764]}")
```

In the context of just identifying what is and isn't a heartbeat, the type of heartbeat doesn't matter so in our eyes the two annotations are not different.
